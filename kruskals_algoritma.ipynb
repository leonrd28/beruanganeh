```python
#algoritma kruskal
# tabel kruskal algoritma kompleks I dan II
import networkx as nx
import matplotlib.pyplot as plt

# Definisikan graf
G = nx.Graph()

# Tambahkan simpul ke graf
G.add_node('21')
G.add_node('1')
G.add_node('2')
G.add_node('3')
G.add_node('4')
G.add_node('5')
G.add_node('6')
G.add_node('7')
G.add_node('8')
G.add_node('9')
G.add_node('10')
G.add_node('11')
G.add_node('12')
G.add_node('13')
G.add_node('14')
G.add_node('15')
G.add_node('16')
G.add_node('17')
G.add_node('18')
G.add_node('19')
G.add_node('20')


# Tambahkan tepi dan bobotnya
G.add_edge('1', '2', weight=6)
G.add_edge('2', '3', weight=6)
G.add_edge('3', '4', weight=6)
G.add_edge('4', '5', weight=6)
G.add_edge('5', '6', weight=6)
G.add_edge('6', '7', weight=6)
G.add_edge('8', '9', weight=6)
G.add_edge('9', '10', weight=6)
G.add_edge('10', '11', weight=6)
G.add_edge('11', '12', weight=6)
G.add_edge('12', '13', weight=6)
G.add_edge('13', '14', weight=6)
G.add_edge('1', '26', weight=6)
G.add_edge('2', '24', weight=6)
G.add_edge('3', '22', weight=6)
G.add_edge('4', '20', weight=6)
G.add_edge('5', '18', weight=6)
G.add_edge('6', '16', weight=6)
G.add_edge('9', '15', weight=6)
G.add_edge('10', '17', weight=6)
G.add_edge('11', '19', weight=6)
G.add_edge('12', '21', weight=6)
G.add_edge('13', '23', weight=6)
G.add_edge('14', '25', weight=6)
G.add_edge('21', '22', weight=8)


# Mencari pohon rentang minimum dengan algoritma Kruskal
minimum_spanning_tree = nx.minimum_spanning_tree(G)

# Tampilkan daftar minimum spanning tree (MST) dengan bobot tepi
print("Minimum Spanning Tree Edges with Weights:")
for u, v in minimum_spanning_tree.edges:
    weight = G[u][v]['weight']
    print(f"{u} - {v} (Weight: {weight}")

# Mencetak bobot total pohon rentang minimum
total_weight = sum([G[u][v]['weight'] for u, v in minimum_spanning_tree.edges])
print("Bobot Total Minimum Spanning Tree:", total_weight)

# Visualisasi pohon rentang minimum
pos = nx.spring_layout(G)  # Layout untuk visualisasi

# Definisikan bobot tepi
edge_labels = {(u, v): d['weight'] for u, v, d in G.edges(data=True)}

# Gambar graf dasar
nx.draw(G, pos, with_labels=True)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels)

# Gambar pohon rentang minimum
nx.draw_networkx_edges(minimum_spanning_tree, pos, edge_color='r', width=2)
nx.draw_networkx_labels(G, pos)

plt.show()
```
